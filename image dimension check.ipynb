```python
import os
from PIL import Image

def get_image_dimensions(folder_path):
    dimensions = set()

    # Iterate over all files in the folder
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        
        # Check if the file is an image
        if os.path.isfile(file_path) and any(filename.endswith(ext) for ext in ['.jpg', '.jpeg', '.png']):
            try:
                # Open the image file
                with Image.open(file_path) as image:
                    # Get the dimensions of the image
                    width, height = image.size
                    dimensions.add((width, height, 3))  # Assuming 3 channels (RGB)
            except:
                print(f"Error processing file: {file_path}")

    return dimensions

# Specify the path to the folder containing the images
folder_path = '/storage/faisal/ISIC2018/training/images'

# Get the dimensions of the images in the folder
image_dimensions = get_image_dimensions(folder_path)

# Print the unique dimensions and count
print(f"Total unique image dimensions: {len(image_dimensions)}")
print("Image dimensions:")
for width, height, depth in image_dimensions:
    print(f"Width: {width}, Height: {height}, Depth: {depth}")
```
